```python
from math import sqrt
def wci(x):
    velocity=x['WindGustSpeed']
    minTemp=x['MinTemp']
    return (10*sqrt(velocity)-velocity+10.5)*(33-minTemp)
```

```python
data['wci']=data.apply(wci,axis=1)
data.head()
```

```python
gold = pd.DataFrame({'Country': ['USA', 'France', 'Russia'],
                         'Medals': [15, 13, 9]}
                    )
silver = pd.DataFrame({'Country': ['USA', 'Germany', 'Russia'],
                        'Medals': [29, 20, 16]}
                    )
bronze = pd.DataFrame({'Country': ['France', 'USA', 'UK'],
                        'Medals': [40, 28, 27]}
                    )
df = gold.merge(silver,how='outer').merge(bronze,how='outer').fillna(0)
df = df.astype({'Medals': float})
df.set_index='Country'
df.groupby(['Country']).sum().sort_values('Medals', ascending=False)
```

```python
df = gold.merge(silver,how='outer').merge(bronze,how='outer').fillna(0)
df['Total']=df.Medals_x+df.Medals_y+df.Medals
df.sort_values('Total',ascending=False)
df=df.drop(['Medals_x','Medals_y','Medals'],axis=1)
df.columns=['Country','Medals']
```

```python
gold_silver=pd.merge(gold, silver, how="outer")
full_ds=pd.merge(bronze, gold_silver, how="outer")
full_ds.fillna(0)
full_ds = full_ds.astype({'Medals': float})
full_ds['Medals'] = pd.to_numeric(full_ds['Medals'])
full_ds = full_ds.groupby(['Country']).sum().sort_values('Medals', ascending=False) #.sort_values(ascending=False)
full_ds
```

```python
def bucket(x):
    if(x >=7.5 and x <8 ):
        return 'A'
    elif(x >=8 and x <8.5 ):
        return 'B'
    elif(x >=8.5 and x <9 ):
        return 'C'
    elif(x >=9 and x <9.5 ):
        return 'D'
    elif(x >=9.5):
        return 'E'
movies['bucket'] =movies.imdb_score.apply(bucket)
movies.groupby('bucket',dropna=True).count()
```
