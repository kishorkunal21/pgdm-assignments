```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('../input/previous-application/previous_application.csv')
data.head()
```

```python
data.shape
```

```python
data.info()
```

There are 3 kinds of data types: int, float, object

```python
data.dtypes.value_counts()
```

```python
data.isnull().sum()
```

We can see that there are some columns having very high number of null values

```python
missing = data.isnull().sum() * 100 / len(data)
missing_df = pd.DataFrame({'columns':data.columns,
                               'percentage missing': missing})
```

```python
missing_df.sort_values('percentage missing',inplace=True, ascending=False)
```

```python
missing
```

```python
columns_to_be_retained = list(missing_df[missing_df['percentage missing']<55].index)
```

```python
data = data[columns_to_be_retained]
```

```python
data.shape
```

```python
sns.countplot(data.NAME_CONTRACT_STATUS)
plt.xlabel('Contract status')
plt.ylabel('Count')
plt.title('Distribution of Contract Status')
plt.show()
```

Most of the loans were approved loans. More loans were canceled than refused.

```python
sns.countplot(data.NAME_CONTRACT_TYPE)
plt.xlabel('Contract type')
plt.ylabel('Count')
plt.title('Distribution of Contract type')
plt.show()
```

Most people opted for CASH LOANS and CONSUMER LOANS

```python
approved = data[data['NAME_CONTRACT_STATUS']=='Approved']
refused = data[data['NAME_CONTRACT_STATUS']=='Refused']
canceled = data[data['NAME_CONTRACT_STATUS']=='Canceled']
unused = data[data['NAME_CONTRACT_STATUS']=='Unused Offer']
```

```python
approved_percentage = (len(approved)*100)/len(data)
refused_percentage = (len(refused)*100)/len(data)
canceled_percentage = (len(canceled)*100)/len(data)
unused_percentage = (len(unused)*100)/len(data)
```

```python
print('Whose loan has been approved',approved_percentage,'%')
print('Whose loan has been refused',refused_percentage,'%')
print('Whose loan has been canceled',canceled_percentage,'%')
print('Whose loan has been unused',unused_percentage,'%')
```
