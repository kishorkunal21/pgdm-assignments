## PCA DEMO

```python
df = pd.read_csv('iris.csv')
```

```python
df.head()
```

```python
df.iris.shape
```

```python
df.iris.value_counts()
```

```python
y = df.pop('iris')
df.head()
```

```python
df.describe()
```

#### Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = df.copy()
x_scaled =  scaler.fit_transform(X)
x_scaled
```

### applying PCA on data

```python
from sklearn.decomposition import PCA
```

```python
?PCA
```

```python
pca = PCA(random_state=42)
```

```python
p = pca.fit_transform(x_scaled)
p
```

```python
pca.components_
```

```python
pca.explained_variance_ratio_
```

```python
pca.explained_variance_
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(range(1, len(pca.explained_variance_ratio_)+1),p#ca.explained_variance_ratio_)
```

```python
#cumulative varience
```

```python
var_cmu = np.cumsum(pca.explained_variance_ratio_)
var_cmu
```

```python
#SCREE PLOT
plt.plot(range(1,len(var_cmu)+1),var_cmu)
```

#### PCA with 2 componenets

```python
pc2 =  PCA(n_components=2, random_state=42)
```

```python
newdata =  pc2.fit_transform(x_scaled)
newdata.shape
```

```python
df_pc2 = pd.DataFrame(newdata, columns=['PC1','PC2'])
df_pc2.head()
```

```python
df_final = pd.concat([df_pc2,y],axis=1)
df_final.head()
```

```python
import seaborn as sns
```

```python
df_final.head()
```

```python
sns.pairplot(df_final, x_vars=['PC1'],y_vars=['PC2'],hue='iris',height=5)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.scatterplot(data=df_final, x='PC1',y='PC2',hue='iris')
plt.show()
```
